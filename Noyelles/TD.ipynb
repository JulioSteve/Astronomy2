```python
from sympy import *

i, M, e, n, eps, th, thdd = symbols(r'i M e n \epsilon \theta \ddot{\theta}')

expr_ar = 1 + 2*Sum(besselj(i,i*e)*cos(i*M),(i,1,8)).doit()
expr_cos = 2*(1-e**2)/e*Sum(besselj(i,i*e)*cos(i*M), (i,1,8)).doit()-e
expr_sin = 2*sqrt(1-e**2)*Sum(diff(besselj(i,i*e)*sin(i*M)/i,e) ,(i,1,8)).doit()


expr = (expr_ar)**3*((2*expr_cos**2-1)*sin(2*th)-2*expr_sin*expr_cos*cos(2*th))
Serie = series(expr, e, 0, 3)
```

```python
trigsimp(Serie)
```

```python
total = thdd + 3/2*n**2*eps*Serie
trigsimp(total)
```
