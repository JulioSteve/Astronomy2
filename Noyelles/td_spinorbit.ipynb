```python
from sympy import *

a, r, n, e, M, eps, theta = symbols(r'a r n e \mathcal{M} \varepsilon \theta', real=True)
i = symbols('i', integer=True)
```

```python
sum_order = 5

a_over_r = 1 + 2*Sum(besselj(i, i*e)*cos(i*M), (i,1,sum_order)).doit()
cos_f = (2*(1-e**2)/e*Sum(besselj(i, i*e)*cos(i*M), (i,1,sum_order))-e).doit()
sin_f = 2*sqrt(1-e**2)*Sum(diff(besselj(i, i*e),e)*sin(i*M)/i, (i,1,sum_order)).doit()
```

```python
second_term = a_over_r**3*((2*cos_f**2 - 1)*sin(2*theta) - 2*sin_f*cos_f*cos(2*theta))
```

```python
series_second_term = series(second_term, e, 0, 3)
```

```python
trigsimp(series_second_term)
```
